## Como obtendrias los eventos, usuarios y origen que contenga el string Failed password for?

```python
!cat $archivo | grep 'Failed password for' | sed 's/.*Failed password for invalid user )//g' | sed 's/.*Failed password for //g' | sed 's/port.*//g' | sed 's/invalid user //g' | sed 's/from //g' > salida_failedpasswd.txt 
```
